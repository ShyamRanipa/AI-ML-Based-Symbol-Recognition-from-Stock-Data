```python
#sum,range,median,mode,variance
import math as ma
print("For vwap column")
n=len(df.iloc[:,10])
sum=0
list=[]
for i in range(n):
  x=int(df.iloc[i,-7])
  list.append(x)
  sum=sum+x
print("Total count is "+str(n))
print("Sum is "+str(sum))
avg=sum/n
print("Average is "+str(avg))
var=0
for j in range(n):
    sd=(list[j]-avg)**2
    var+=sd
var=var/(n-1)
st1=ma.sqrt(var)
print("Variance of given data is "+str(var))
print("Standard Deviation of given data is "+str(st1))
```
